```python
import numpy as np
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

#Membaca file Excel DataTA.xlsx
df = pd.read_excel('DataTA.xlsx')

#Membaca kolom Waktu Antar Kedatangan (detik) dari file Excel
data = df['WAKD'].values
print("Data Waktu Antar Kedatangan:")
print(data)
print('')

# Uji Kolmogorov-Smirnov untuk distribusi normal
# Normalisasi data
data_normalized = (data - np.mean(data)) / np.std(data)
print("Rata-rata: ",np.mean(data))
print("Standar Deviasi: ",np.std(data))
ks_stat_normal, p_value_normal = stats.kstest(data_normalized, 'norm')

print(f'K-S Test Statistic (Normal): {ks_stat_normal}, p-value: {p_value_normal}')

if p_value_normal > 0.05:
    print('Data berdistribusi normal berdasarkan uji Kolmogorov-Smirnov')
else:
    print('Data tidak berdistribusi normal berdasarkan uji Kolmogorov-Smirnov')
print('')

# Uji Kolmogorov-Smirnov untuk distribusi Poisson
# Estimasi parameter lambda (mean dari data)
lambda_poisson = np.mean(data)
print("λ poisson: ", lambda_poisson)

#Melakukan Uji Kolmogorov-Smirnov
ks_stat_poisson, p_value_poisson = stats.kstest(data, 'poisson', args=(lambda_poisson,))

print(f'K-S Test Statistic (Poisson): {ks_stat_poisson}, p-value: {p_value_poisson}')

if p_value_poisson > 0.05:
    print('Data berdistribusi Poisson berdasarkan uji Kolmogorov-Smirnov')
else:
    print('Data tidak berdistribusi Poisson berdasarkan uji Kolmogorov-Smirnov')
print('')
    
# Uji Kolmogorov-Smirnov untuk distribusi Seragam
# Rentang data
min_val = np.min(data)
max_val = np.max(data)
print("Data Minimum: ",min_val)
print("Data Maksimum: ",max_val)

#Melakukan Uji Kolmogorov-Smirnov
ks_stat_uniform, p_value_uniform = stats.kstest(data, 'uniform', args=(min_val, max_val - min_val))

print(f'K-S Test Statistic (Uniform): {ks_stat_uniform}, p-value: {p_value_uniform}')

if p_value_uniform > 0.05:
    print('Data berdistribusi Seragam berdasarkan uji Kolmogorov-Smirnov')
else:
    print('Data tidak berdistribusi Seragam berdasarkan uji Kolmogorov-Smirnov')
print('')
    
# Uji Kolmogorov-Smirnov untuk distribusi Eksponensial
# Estimasi parameter lambda (1/mean dari data)
lambda_expon = 1 / np.mean(data)
print("λ exponent: ", lambda_expon)

#Melakukan Uji Kolmogorov-Smirnov
ks_stat_expon, p_value_expon = stats.kstest(data, 'expon', args=(0, 1/lambda_expon))

print(f'K-S Test Statistic (Exponential): {ks_stat_expon}, p-value: {p_value_expon}')

if p_value_expon > 0.05:
    print('Data berdistribusi Eksponensial berdasarkan uji Kolmogorov-Smirnov')
else:
    print('Data tidak berdistribusi Eksponensial berdasarkan uji Kolmogorov-Smirnov')
    
# QQ plot
sm.qqplot(data, line='s')
plt.title('Plot Uji Distribusi WAKD')
plt.show()
```
